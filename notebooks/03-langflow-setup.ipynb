## ⚙️ Step 5: Configure Flow Components

After importing, the flow components should already be pre-configured. Please double-check the settings for each component below. The **only component you need to configure** is the LLM component — the credentials will be provided by your instructor.

### Configuration Checklist:

#### 1. **Token Exchange Tool** ✅ Pre-configured — please double-check
   - Click on the Token Exchange component
   - Verify the following parameters are set:
     - `token_endpoint`: Your IBM Verify token endpoint
     - `client_id`: Your OAuth client ID
     - `client_secret`: Your OAuth client secret
     - `actor_token`: Your actor token

#### 2. **Vault Credentials Tool** ✅ Pre-configured — please double-check
   - Click on the Vault component
   - Verify the following are set:
     - `vault_url`: `http://vault:8200` (or your Vault URL)
     - `vault_token`: Your Vault token (from previous notebooks)
     - `role_name`: `hr-basic` or `hr-admin`

#### 3. **Database Tool** ✅ Pre-configured — please double-check
   - Click on the Database component
   - Verify connection settings:
     - `host`: `postgres` (Docker service name)
     - `port`: `5432`
     - `database`: `hr_database`
     - Note: Credentials will be provided dynamically by Vault

#### 4. **LLM Component** ⚠️ Requires configuration
   - Click on the LLM component
   - Select **IBM watsonx.ai** as the model provider
   - Configure the following (credentials provided by your instructor):
     - `Project ID`: Your watsonx.ai project ID
     - `Endpoint URL`: Your watsonx.ai endpoint URL
     - `API Key`: Your watsonx.ai API key

> 📝 **Note:** In actual production deployments, the LLM API key should also be stored and retrieved securely from Vault — just like the database credentials. Storing secrets directly in the UI is not recommended for production use. This is out of scope for this workshop, but worth keeping in mind.

### 📋 Viewing Agent Logs in Langflow

Even without running the flow from the playground, Langflow exposes detailed logs for each component after a flow execution. Once you have tested the flow through the frontend, you can come back here to inspect what happened under the hood:

- **Agent logs**: See the reasoning steps the agent took and which tools it decided to call
- **Tool outputs**: View the output returned by each tool (Token Exchange, Vault, Database)
- **LLM input/output**: Inspect the prompt sent to the model and the response generated

To view logs:
1. Click on any component in the flow canvas
2. Look for the **Logs** or **Output** panel on the component
3. Review the execution trace from the last run

> 💡 After testing through the frontend in the next notebook, come back to Langflow and review the agent logs to see exactly what the agent did — which tools it called, what data was retrieved, and how the final response was constructed.

## 🔗 Step 7: Get Flow API Endpoint

Once your flow is working, you can get the API endpoint for integration:

### Instructions:

1. **Open Flow Settings**
   - Click the **settings icon** or **menu (⋮)** in the flow
   - Look for **"API"** or **"Endpoint"** section

2. **Copy API Endpoint**
   - You should see an endpoint URL like:
     ```
     http://localhost:7860/api/v1/run/{flow_id}
     ```
   - Copy this URL for use in your application

3. **Get API Key** (if not already done)
   - Go to **Settings** → **API Keys**
   - Create a new API key
   - Save it securely in your `.env` file
